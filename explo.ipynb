```python
from typing import List
import json 
import requests

def construct_link(coordinates: List) -> str:
    base_link = "https://apicarto.ign.fr/api/wfs-geoportail/search?source=BDTOPO_V3%3Abatiment&geom=%7B%22coordinates%22%3A%5B%5B%5B"
    if len(coordinates) == 5:
        print("SImpke ploygon")
        end_link = "%5D%5D%5D%2C%22type%22%3A%22Polygon%22%7D"
        for nb_point in range(len(coordinates)):
            if nb_point != (len(coordinates)-1):
                base_link += f"{coordinates[nb_point][0]}%2C{coordinates[nb_point][1]}%5D%2C%5B"
            else : 
                 base_link += f"{coordinates[0][0]}%2C{coordinates[0][1]}"
        final_link = base_link + end_link
        # print(final_link)
        # return(final_link)
    else : 
        base_link += "%5B"
        print("Multipolygon")
        end_link = "%5D%5D%5D%5D%2C%22type%22%3A%22MultiPolygon%22%7D"
        for nb_point in range(len(coordinates)):
            if nb_point != (len(coordinates)-1):
                base_link += f"{coordinates[nb_point][0]}%2C{coordinates[nb_point][1]}%5D%2C%5B"
            else : 
                 base_link += f"{coordinates[0][0]}%2C{coordinates[0][1]}"
        final_link = base_link + end_link
    return final_link


def get_buildings(coordinates : List) -> str :
    link = construct_link(coordinates)
    response = requests.get(link)
    try :

        with open('bd_topo.geojson', 'w') as f:
            json.dump(response.json(), f)
        print("File saved successfully")
    except :
        print(f"Error : HTML code = {response}. Verify the link")

get_buildings(coordinates)
```
